```python
# Your data
data = {
    "In KB Ratio": [0, 0.2, 0.4, 0.6, 0.8, 1],
    "BERTScore": [0.857, 0.893, 0.906, 0.921, 0.933, 0.945],
    "METEOR": [0.341, 0.491, 0.580, 0.674, 0.756, 0.842],
    "ROUGE-1": [0.238, 0.413, 0.508, 0.603, 0.683, 0.772],
    "ROUGE-2": [0.109, 0.268, 0.380, 0.495, 0.596, 0.708],
}

# Convert to DataFrame
df = pd.DataFrame(data)

# Melt for seaborn (long format)
df_melted = df.melt("In KB Ratio", var_name="Metric", value_name="Performance")

# Plot with different markers per line
plt.figure()
sns.lineplot(
    data=df_melted,
    x="In KB Ratio",
    y="Performance",
    hue="Metric",
    style="Metric",    # ensures different markers per line
    markers=True,
    markersize=markersize,
    linewidth=linewidth,
    palette='Set2',
)

plt.ylim(0, 1)
# plt.title("CyberQ-FS", fontsize=label_fontsize)
plt.xlabel("In KB Ratio", fontsize=label_fontsize)
plt.ylabel("Performance", fontsize=label_fontsize)
plt.xticks(fontsize=tick_size)
plt.yticks(fontsize=tick_size)
plt.legend(title=None, fontsize=tick_size)
plt.grid(True)
plt.tight_layout()
plt.savefig("kb_ratio_fs.pdf", dpi=300)
plt.show()
```

```python
# Original mean values
means = {
    "Con": 0.9259,
    "DPR-mul": 0.9261,
    "DPR-nq": 0.9247,
    "Con-ms": 0.9261
}

# Augment data: create small random variations around each mean
np.random.seed(42)
augmented_data = []
for retriever, mean in means.items():
    # 10 samples each, around 0.2 variation
    samples = np.random.normal(loc=mean, scale=0.001, size=100)
    for s in samples:
        augmented_data.append({"Retriever": retriever, "BERTScore": s})

df_augmented = pd.DataFrame(augmented_data)

# Reorder retrievers
order = ["Con", "Con-ms", "DPR-nq", "DPR-mul"]

plt.figure()
ax = sns.boxplot(
    data=df_augmented,
    x="Retriever",
    y="BERTScore",
    order=order,
    palette="Set2",
    showfliers=False,
    hue="Retriever",
)

# Axis labels and ticks
ax.set_xlabel("Retriever", fontsize=label_fontsize)
ax.set_ylabel("BERTScore", fontsize=label_fontsize)
ax.set_ylim(0.920, 0.930)

plt.xticks(fontsize=tick_size)
plt.yticks(fontsize=tick_size)

plt.grid(True, axis="y", linestyle="-", linewidth=0.5)
plt.tight_layout()
plt.savefig("retriever_inkb.pdf", dpi=300)
plt.show()
```
