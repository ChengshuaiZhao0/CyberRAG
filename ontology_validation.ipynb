```python
from together import Together
```

```python
TOGETHER_API_KEY = "Your API Key Here"
ontology_path = 'dataset/ontology/'
```

```python
def get_response(client, input_text):
    response = client.chat.completions.create(
    model="meta-llama/Llama-3.3-70B-Instruct-Turbo-Free", #cs: here is a free api provide by together AI, I am not sure how many credits we can use.
    messages=[
      {
        "role": "system", 
        "content": "You are CyberCloud AI Buddy...."}, #cs: add some information about our project here.
      {
        "role": "user", 
        "content": input_text},
    ],
    )
    return response
```

```python
def answer_check(answer):
    answer = answer.lower()
    reject_words=['no','not','none','nothing','nope','nada','nah','never','negative','n','nil','nix','nought','zero','zilch','zip','zippo','zot','zotz']
    if answer in reject_words:
        return False
    else:
        return True
```

```python
def answer_filter(answer, respose_content):
    if answer_check(respose_content):
        return answer
    else:
        return respose_content
```

```python
def ontology_validation(question, answer):
    with open(ontology_path + 'ontology.txt', 'r') as f:
        ontology_text = f.read()
    # cs: provide question user queried and answer llm generated, here provide an example:
    question = "In what way can decryption assist in detecting insecure encryption implementations?"
    answer = "ANSWER: According to the DOCUMENT, decryption can assist in detecting insecure encryption implementations by helping to identify vulnerabilities in the encryption process that could be exploited by attackers. In other words, decryption can be used to analyze the encryption process and identify any weaknesses or flaws that could be exploited by malicious actors. This can help to improve the overall security of the system by identifying and addressing potential vulnerabilities before they can be exploited."

    ontology = ontology_text
    instruction = 'Please judge if the QUESTION and ANSWER align well with the ONTOLOGY.\nThe QUESTION and ANSWER align well with the ONTOLOGY if the QUESTION and ANSWER are in the same knowledge domain as the ONTOLOGY.\nThe output format is a tuple: (your judgment: Pass/Not Pass,  confidence score)'
    input_text = 'QUESTION:\n{0}\n\nANSWER:\n{1}\n\nONTOLOGY:\n{2}\n\nINSTRUCTIONS:\n{3}'.format(question, answer, ontology, instruction)
    client = Together(api_key = TOGETHER_API_KEY)
    response = get_response(client, input_text)
    out = answer_filter(answer, response.choices[0].message.content)
    return out
```
